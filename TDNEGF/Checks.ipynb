```julia
using Distributions
N_θ = 100
Δθ = pi
J=10.
h=0.0
function P(θ,J,h) #### Probability distribution
    m = 0.9
    exp((J*m +h)*cos(θ))*(J*m+h)/(sinh(J*m+h)*2)
end
using QuadGK
#probs = [P(θ,10,0) for θ in range(0,Δθ,N_θ)]
norm, err = quadgk(θ -> P(θ, J, h), 0.0, π)
P(θ) =  P(θ,J,h)
norm_P(θ) = P(θ, J, h) / norm
cdf_P(θ) = quadgk(norm_P,0.0,θ)[1]


function discretize_pdf(pdf; θ_min, θ_max, bins)#, J, h)
    bin_edges = range(θ_min, θ_max, length=bins+1)
    bin_probs = zeros(bins)
    dθ =(θ_max-θ_min)/(bins)
    #### This is the legal way to get it
    for i in 1:bins
        #bin_probs[i], _ = quadgk(θ -> pdf(θ, J, h), bin_edges[i], bin_edges[i+1])
        bin_probs[i] = (pdf(bin_edges[i], J, h) + pdf(bin_edges[i+1], J, h))*dθ/2
    end
    
    
    # Normalize the probabilities to ensure they sum to 1
    bin_probs /= sum(bin_probs)
    return bin_edges, bin_probs
end
bin_edges, bin_probs = discretize_pdf(P; θ_min=0., θ_max=pi, bins=1000)#, J, h)
# Create a categorical distribution with the bin probabilities
cat_dist = Categorical(bin_probs)
# Function to sample from the discretized distribution
function samples(cat_dist, bin_edges, num_samples)
    samples = rand(cat_dist, num_samples)
    mid_points = (bin_edges[1:end-1] + bin_edges[2:end]) / 2
    return mid_points[samples]
end
# # Generate samples
# sample = samples(cat_dist, bin_edges, 10)
# # using Plots
# histogram(sample, bins=20, label="Samples check")


#Categorical(probs)
# function sample(N::Int)
#     samples = Float64[]
#     while length(samples) < N
#         θ_r = rand(Uniform(0, π))
#         u = rand()
#         pr = norm_P(θ_r)
#         maxi = maximum(pr)
#         if u < pr/maxi#cdf_P(θ_r) #norm_P(θ_r) / (1/(π))  # Assuming the proposal density is uniform
#             push!(samples, θ_r)
#         end
#     end
#     return samples
# end
#samples = sample(10)/pi
# θs=range(0,pi,100)
# pr = norm_P.(θs)
# maxi = maximum(pr)
# plot(pr/maxi)
# plot!(cdf_P.(θs))
```

```julia
function rand_init_vecs!(dynamics_var)
    for jj in 1: global_var.n::Int  ### Run over the number of sites of the lattice 
        ϕ = rand(Uniform(0, 2π))#(rand()-0.5)*0.1
        θ = samples(cat_dist, bin_edges, global_var.n)[jj]#sample(global_var.n)[jj]#samples[jj]
        dynamics_var.vm_a1x[jj][1] = sin(θ)*cos(ϕ)#0.9 #0.1 + (rand()-0.5)*0.1
        dynamics_var.vm_a1x[jj][2] = sin(θ)*sin(ϕ) #(rand()-0.5)*0.1
        dynamics_var.vm_a1x[jj][3] = cos(θ)#(rand()-0.5)*0.1#0.0 #0.9#(-1)^(jj+1)#1.
    end
    return nothing
end
### Average over many experiment 
function avrg_exp(N_exp;tf=500,dynamics_var=dynamics_var)
    mag_avg =zeros(N_exp,Int(tf/0.5),3*global_var.n)
    #Matrix{Float64}(undef,tf*10,3*global_var.n)
    for ni in range(1,N_exp)
        rand_init_vecs!(dynamics_var)
        evol_llg(0.5,0.5,tf,dynamics_var)
        mag = readdlm("./data/test_jl.txt")[:,2:end]#mag#[:,site*3+1]
        #mag_avg += mag
        mag_avg[ni,:,:] = mag
        #push!(mag_avg, mag)
        println("Experiment:", ni)
    end
    #mag_avg/N_exp ;
    mag_avg
end
```
